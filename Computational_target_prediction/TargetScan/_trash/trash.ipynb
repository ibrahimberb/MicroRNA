```python
def count_overlapping(string, sub_string):
    count = 0
    for pos in range(len(string)):
        if string[pos:].startswith(sub_string):
            count += 1
    return count
```

```python
from typing import List
from pandas import DataFrame


def run(input_genes: List[str], pred_target_info_data: DataFrame, gene_mapping: dict, UTR_seq_data: DataFrame):
    entries = []
    for gene in input_genes:
        query = pred_target_info_data[
            pred_target_info_data["Gene ID"].apply(lambda x: x.split(".")[0]) == gene
            ]

        sample_utr_seqs = get_UTR_seq_from_ENSMUSG(UTR_seq_data, gene)
        sample_utr_seqs_nondash = get_UTR_seqs_nondash(sample_utr_seqs)

        entries.append(
            (gene, gene_mapping[gene], len(query), sample_utr_seqs_nondash)
        )

    return pd.DataFrame(
        entries, columns=["GENE_ENSMUSG", "GENE_SYMBOL", "#_SITE", "UTR_SEQS"]
    )
```
